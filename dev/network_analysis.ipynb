# Description

End-to-end network analysis pipeline, incl. data onboarding.

This pipeline reads network graph data from a csv file that contains edges data, as well as an optional csv file containing node attributes (and/or nodes that are not connected to other nodes.)

It then converts this data into a network graph structure, and extracts graph properties (like density, number of edges, number of nodes, ...) and graph component properties (largest component, number of components).


## Connecting the workflow steps <a class="anchor" id="connecting_steps"></a>

After we all steps are created, we need to wire them up to create our workflow structure. All step inputs that are not
connected to other step outputs are considered user inputs, and, if their values are marked as 'required', must be set
later in order for our workflow to run successfully.

#### Connecting inputs for: ``load_edges_table``

```python
step_load_edges_table.input.file = step_read_edges_file.output.file
```

#### Connecting inputs for: ``load_nodes_table``

```python
step_load_nodes_table.input.file = step_read_nodes_file.output.file
```

#### Connecting inputs for: ``create_graph``

```python
step_create_graph.input.edges_table = step_load_edges_table.output.table
```

#### Connecting inputs for: ``augment_graph``

```python
step_augment_graph.input.graph = step_create_graph.output.graph
```

```python
step_augment_graph.input.node_attributes = step_load_nodes_table.output.table
```

#### Connecting inputs for: ``extract_graph_properties``

```python
step_extract_graph_properties.input.graph = step_augment_graph.output.graph
```

#### Connecting inputs for: ``extract_largest_component``

```python
step_extract_largest_component.input.graph = step_augment_graph.output.graph
```
